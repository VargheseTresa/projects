### Establishing connection

```python
import sqlite3
from sqlite3 import Error
import pandas as pd
```

```python
try:
    conn = sqlite3.connect("sqlite_db_pythonsqlite.db")
    cur = conn.cursor()
except Error as e:
        print(e)
        
def execute_fetch(query):
    cur = conn.cursor()
    cur.execute(query)
    result = cur.fetchall()
    
    return result

```

#### Q10: Produce a list of facilities with a total revenue less than 1000. The output of facility name and total revenue, sorted by revenue.

```python
q = """
    SELECT  
        name,
        SUM(CASE WHEN Bookings.memid = 0 THEN slots * guestcost
                ELSE slots * membercost END) AS TotalRevenue
    FROM Bookings 
        INNER JOIN Facilities ON Bookings.facid = Facilities.facid
    GROUP BY name
    HAVING TotalRevenue < 1000
    ORDER BY TotalRevenue
    """
pd.DataFrame(execute_fetch(q), columns=['Facility Name', 'Total Revenue'])
```

#### Produce a report of members and who recommended them in alphabetic surname,firstname order

```python
q = """
    SELECT 
        m1.surname AS Member_surname,
        m1.firstname AS Member_firstname,
        m2.surname AS RecommendedBy_Surname,
        m2.firstname AS RecommendedBy_Firstname
    FROM Members AS m1
        INNER JOIN Members as m2 ON m1.recommendedby = m2.memid
    WHERE m1.recommendedby > 0
    ORDER BY m2.surname, m2.firstname
    """
pd.DataFrame(execute_fetch(q), columns=['Member_surname', 'Member_firstname', 'RecommendedBy_Surname', 'RecommendedBy_Firstname'])
```

#### Find the facilities with their usage by member, but not guests

```python
q = """
    SELECT name, sum(slots) AS Total_slots_booked 
    FROM Bookings
        INNER JOIN Facilities ON Facilities.facid = Bookings.facid 
    WHERE memid > 0
    GROUP BY name
    ORDER BY Total_slots_booked
    """

pd.DataFrame(execute_fetch(q), columns=['Facility Name', 'Total slots booked'])
```
